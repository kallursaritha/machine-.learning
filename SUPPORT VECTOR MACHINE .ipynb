# Data Visualization

```python
sns.heatmap(iris.corr(),annot=True )
```

```python
iris1 = iris[iris['species']=='Iris-setosa']
iris2 = iris[iris['species']=='Iris-versicolor']
iris3 = iris[iris['species']=='Iris-virginica']

plt.scatter(iris1['petal_length'],iris1['petal_width'], color='r' , label='Iris-setosa')
plt.scatter(iris2['petal_length'],iris2['petal_width'], color='b', label='Iris-versicolor')
plt.scatter(iris3['petal_length'],iris3['petal_width'], color='g' , label='Iris-virginica ')


plt.legend()
plt.show()
     
```

```python
iris_imp = iris.iloc[:,0:4]
from sklearn.cluster import KMeans
k_meansclus = range(1,10)
sse = []

for k in k_meansclus :
  km = KMeans(n_clusters =k)
  km.fit(iris_imp)
  sse.append(km.inertia_)
```

```python
sse
```

```python
plt.title('The Elbow Method')
plt.plot(k_meansclus,sse)
plt.show()
```

```python
km1 = KMeans(n_clusters=3,max_iter=300 , random_state=0)
km1.fit(iris_imp)
     
```

```python
y_means = km1.fit_predict(iris_imp)
y_means
     
```

```python
km1.cluster_centers_
```

```python
iris_imp = np.array(iris_imp)
iris_imp
```

```python
plt.scatter(iris_imp[y_means==1,2 ],iris_imp[y_means==1,3 ], color='r' , label='Iris-setosa')
plt.scatter(iris_imp[y_means==2,2 ],iris_imp[y_means==2,3 ], color='b', label='Iris-virginica')
plt.scatter(iris_imp[y_means==0,2 ],iris_imp[y_means==0,3 ], color='g' , label='Iris-versicolor ')



plt.legend()
plt.show()
     
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
# Separating the independent variables from dependent variables
x=iris.iloc[:,:-1]
y=iris.iloc[:,4]
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.30)
```

# Training and Fitting the model

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train, y_train)
```

# Predictions from the trained model

```python
pred=model.predict(x_test)
```

# Model Evaluation

```python
# Importing the classification report and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test, pred))
```
